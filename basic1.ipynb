```python
# Problem 1

# Loading data
biodata = csv.reader(open('biocube_data.csv','r'))
# Creating empty list
data = []

# Placing and cleaning data into empty list
for row in biodata:
    data.append(row)

del data[0:2]
```

```python
# Empty list
company1 =[]
company2 =[]
company3 =[]

# Seperating data into company lists
for i in range(len(data)):
    company1.append(int(data[i][1]))
    company2.append(int(data[i][2]))
    company3.append(int(data[i][3]))
```

```python
#Problem 2 Question 11
# Histogram

# Number of rolls and empty vector
roll =1000000
vector =[]

# Rolling random numbers between 1 and 6. Placing values into a vector
while roll != 0:
    roll -=1
    vector.append(random.randint(1,6))

# Plotting the histograms
plt.hist(vector,bins =[0.5,1.5,2.5,3.5,4.5,5.5,6.5],align = 'mid', rwidth = 0.5)
plt.xlabel('Face Rolled', labelpad = 20)
plt.ylabel('Frequency', labelpad = 20)
plt.title("Outcomes of the Frequency of a One-Six Sided Die Roll 1 Million Times")
plt.show()
```

```python
plt.hist(vector,bins =[0.5,1.5,2.5,3.5,4.5,5.5,6.5],align = 'mid', rwidth = 0.5,density=True)
plt.xlabel('Face Rolled', labelpad = 20)
plt.ylabel('Chances', labelpad = 20)
plt.title("Probability Density Function of a One-Six Sided Die Roll 1 Million Times")
plt.show()
```
