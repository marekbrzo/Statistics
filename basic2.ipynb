PART 1:

Legojoy is a chain of toy stores. They sell lots of different kinds of toys for kids. Recently, they are facing tremendous competition from online stores and other selling channels. They need to find a way to thrive.
One of the suggestion is that they have to create a competitive advantage in term of enhancing customer experiences (customers are mostly kids and parents). They implement a proof of concept to use augmented reality in the stores so that kids and parents can plays with toys in 3D without opening the box. Now, 19 stores have implemented this new concept and collected information of average daily profit per customer per toy. Legojoy now wants to see if the new concept has any effect within 95% confidence level.


```python
# Part one Parameters & Figures
meanToy = 20.5
stdToy = 9
popToy = 130

# Figure to produce normalized Z value
range = np.arange(-3,3,0.001)
def draw_z_score( x, cond, mean, std, title):
    y = norm.pdf(x,mean,std)
    z = x[cond]
    plt.plot(x,y)
    plt.fill_between(z, 0 , norm.pdf(z,mean,std))
    plt.title(title)
    plt.show()
```

Question 1
Calculate alpha.

```python
CI = 95
alpha = (100 - CI)/100
twoTailAlpha = alpha/2

print('For a one tail we have an alpha of: ', alpha)
print('For a two tail we have an alpha of: ', twoTailAlpha)
```

Since we are asked to find any change to our sales we MUST look at our two tail alpha, which is 2.5% or 0.025

Question 2 Calculate z

```python
z = abs(norm.ppf(twoTailAlpha))
print('The z-value for a normal two-tail distribution with a confidence of 95 percent is:', round(z,2))

draw_z_score(range,(-z < range) & (range < z),0,1,'Two Tail Confidence of 95 percent')
```

Question 3 Calculate the lower boundary for a 90% confidence interval

```python
CI = 90
alpha = (100 - CI)/100
twoTailAlpha = alpha/2

print('For a two tail we have an alpha of: ', twoTailAlpha)

z = abs(norm.ppf(twoTailAlpha))
print('The z-value for a normal two-tail distribution with a confidence of 90 percent is:', round(z,2))
```

```python
boundary = meanToy - (z * stdToy) / popToy**(1/2)
draw_z_score(range, (range < -z),0,1,'Lower Tail of a 90 percent Confidence')
print('The lower boundary for a 90 percent condifence interval is: ', round(boundary,2))
```

Question 4 Calculate the upper boundary for a 90% confidence interval

```python
boundary = meanToy + (z * stdToy) / popToy**(1/2)
draw_z_score(range, (range > +z),0,1,'Upper Tail of a 90 percent Confidence')
print('The upper boundary for a 90 percent condifence interval is: ', round(boundary,2))
```

Part 2:

The previous average sale of education toy was 45.31 dollar. Now, you want to construct a hypothesis test to test whether the the average spend of clients on this toy changes or not . We collect data of a sample of 240 sales which normally distributed with a mean of  21 and a standard deviation of 11 dollar. 


```python
# Part two Parameters & Figures
previousSale = 45.31
meanSale = 21
stdSale = 11
popSale = 240
```

Question 5.

Please construct the hypothesis to test if the new concept stores generate any DIFFERENCE in sales from the old test

Answer:


FOR OUR TEST

Null Hypothesis: 

Average spend of clients on this toy has NOT CHANGED from previous average clients spend   (H_0 : Sale_new EQUALS Sale_Old)

Alternative Hypothesis: 

Average spend of clients on this toy has  CHANGED from previous average clients spend   (H_0 : Sale_new DOES NOT EQUAL Sale_Old)

'We may never actually 'accept' the NULL hypothesis, only choose to not reject it.'

Question 6. 

Is it one or two tail test?

Answer: 

When measuring ANY DIFFERENCE or IS NOT THE SAME AS we must use a TWO TAILED TEST. In our problem we are asked to determine if the mean (average sale) has translated to the left or right. A significant shift in either direction ourside our confidence level will help us reject our Null hypothesis.

Question 7. Calculate test statisitics c.

Test Statistic for One Population Mean = ( mean_sample - previous_mean ) * sqrt(population_sample) / std_sample


```python
c = abs(meanSale - previousSale)*((popSale)**(1/2)/stdSale)
print(' Test statistics c is: ' , round(c,2))
```

Question 8. Calculate p-value of the new sample.

P-value: The probability, under the assumption of hypothesis of obtaining a result equal to or more extreme than what was actually observed

```python
# Two tail p-value calculations.
p_value = 2*(1-norm.cdf(c))
print("The p-value is equal to:", p_value)
```

The smaller the p-value, the higher the evidence against the null hypothesis, so we may reject the null hypothesis. There is a decrease in sale with the newer strategy.
